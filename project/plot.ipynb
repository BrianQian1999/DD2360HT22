```python
import matplotlib.pyplot as plt
import numpy as np

name_list = ['CPU Memory', 'Paged Memory', 'Pinned Memory', 'Unified Memory']

mov_time = [3.03502,0.0690401,0.00155332,0.01541]
int_time = [2.27175,2.21246,0.0336429,2.30392]


bar_width = 0.3
x_mov = np.arange(len(name_list))
x_int = x_mov + bar_width

plt.bar(x_mov, height=mov_time, width=bar_width, color='g', label="Mover Time")
plt.bar(x_int, height=int_time, width=bar_width, color='b', label="Interpolation Time")

plt.legend()
plt.xticks(x_mov+bar_width/2, name_list)
plt.xlabel("Type of Memory")
plt.ylabel("Time/Cycle (s)")
plt.title("Mover Time and Interpolation Time of Different Memory Systems with 2D Input")

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

name_list = ['Pinned Memory', 'Unified Memory']

HtoD= [1.8000,11.07200]
DtoH = [1.7150,7.968000]


bar_width = 0.3
x_HtoD = np.arange(len(name_list))
x_DtoH = x_HtoD + bar_width

plt.bar(x_HtoD, height=HtoD, width=bar_width, color='g', label="Host to Devvice")
plt.bar(x_DtoH, height=DtoH, width=bar_width, color='b', label="Device to Host")

plt.legend()
plt.xticks(x_HtoD+bar_width/2, name_list)
plt.xlabel("Type of Memory")
plt.ylabel("Time (us)")
plt.title("Data Transfer of Different Memory Systems with 2D Input")

plt.show()
```
